## Exploratory Data Analysis

### 1. Which genres has most popularity each year?

```python
year1 = set()
genre1 = set()
genre_year = movies_data[["genres", "release_year"]]
```

```python
#set of inimitable years to be created
year_of_production = genre_year["release_year"]
year_of_production = year_of_production.drop_duplicates()
for year in year_of_production:
    if year not in year1:
        year1.add(year)
```

```python
print(year1)
```

```python
#set of inimitable genres to be created
for year in year1:
    genre_info = {}
    yearly_genres = genre_year[genre_year.release_year == year]
    yearly_genres = yearly_genres["genres"].values
    for elem in yearly_genres:
        genresrow= elem.split("|")
        for genre in genresrow:
            if genre not in genre1:
                genre1.add(genre)
```

```python
print(genre1)
```

```python
#dataframe with summing the movies yearly
gerne_yearly_count = pd.DataFrame(index = year1, columns=genre1)
gerne_yearly_count[:] = 0  
```

```python
for year in year1:
    genre_info = {}
    yearly_genres = genre_year[genre_year.release_year == year]
    yearly_genres = yearly_genres["genres"].values
    for elem in yearly_genres:
        genresrow = elem.split("|")
        for genre in genresrow:
            if genre not in genre_info:
                genre_info[genre] = 1
            else:
                genre_info[genre] = genre_info[genre] + 1
                    
    aux_df = pd.DataFrame(genre_info, index = [year])
    gerne_yearly_count.loc[year, aux_df.columns] = gerne_yearly_count.loc[year, aux_df.columns] + aux_df.loc[year]
```

```python
print(gerne_yearly_count.apply( max, axis=1 ))
print(gerne_yearly_count.idxmax(axis = 1))
```

```python
#popular geres in various movies yearly
popular_genre_yearly = pd.DataFrame([gerne_yearly_count.idxmax(axis = 1).values,
                                          gerne_yearly_count.apply( max, axis=1 ).values],
                                          columns = gerne_yearly_count.index,
                                         index = ["genre", 'counts'])
```

```python
popular_genre_yearly
```

### From the table above 
In the first five years fromthe dataset we notice that drama was the most popular in 1960,1961,1962 and 1964 but in 1963 thriller took over as the most popular movie that year.
In the last five years from the data set we notice that drama was the most popular movie in from 2010 until 2015.
